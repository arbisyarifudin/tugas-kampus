##Mencetak Tipe Data
**Praktik 2.4.1**
Cobalah mengetik kode di bawah ini :
``` javascript
#tipe data Boolean
print(True)


#tipe data String
print("Mari belajar Python")
print('Belajar Python Sangat Dengan Mudah')


#tipe data Integer
print(20)


#tipe data Float
print(3.14)


#tipe data List
print([1,2,3,4,5])
print(["satu", "dua", "tiga"])


#tipe data Tuple
print((1,2,3,4,5))
print(("satu", "dua", "tiga"))


#tipe data Dictionary
print({"nama":"Budi", 'umur':20})

#tipe data Dictionary dimasukan ke dalam variabel biodata
biodata = {"nama":"Andi", 'umur':21} #proses inisialisasi variabel biodata
print(biodata) #proses pencetakan variabel biodata yang berisi tipe data Dictionary
type(biodata) #fungsi untuk mengecek jenis tipe data. akan tampil <class 'dict'>
``` 

````
#Jawab Praktik 2.4.1

# ini tipe data boolean
print(True) 

# ini tipe data string (harus dalam petik)
print("Belajar bahasa python yuk!")
print("Belajar python itu mudah.")

# ini tipe data integer (bilangan bulat)
print(21) # nomornya eyeshield sih inimah

# ini tipe data float (bilangan pecahan)
print(3.14) # ini Pi

# ini tipe data list
print(1,2,3,'empat',5)
print('tempe', 'tahu', 'nasi', 'kerupuk', 'sambel') # mantapp
# btw kalau dalam bahasa pemprograman lain; LIST itu mungkin sama dengan ARRAY yah?

# tipe data tuple (nilainya tidak bisa diubah)
print(1,2,3,4,5)
print('satu', 'dua', 3, 'empat')

# ini tipe dictionary
print({"nama":"Arbi Syarifudin", "Umur": 25}) # loh kok mirip JSON? hehe

#tipe data Dictionary dimasukan ke dalam variabel myBiodata
myBiodata = {"nama":"Arbi Syarifudin", 'Umur':25} #proses inisialisasi variabel myBiodata
print(myBiodata) #proses pencetakan variabel myBiodata yang berisi tipe data Dictionary
type(myBiodata) #fungsi untuk mengecek jenis tipe data. akan tampil <class 'dict'>

````

**Tugas Praktek 3.3**

1.Printing Data type menggunakan Variable.
Ketikan kode di bawah ini pada Code Editor.
``` javascript
var_string="Mari Belajar Python"
var_int=10
var_float=3.14
var_list=[1,2,3,4]
var_tuple=("satu","dua","tiga")
var_dict={"nama":"Ali", 'umur':20}


print(var_string)
print(var_int)
print(var_float)
print(var_list)
print(var_tuple)
print(var_dict)
```


````
#Jawab Tugas Praktek 3.3 Nomor 1

# mendefinisikan variabel dan isinya
varString = "Mari Belajar Python"
varInt = 33
varFloat = 9.8
varList = [1,2,'tiga',4]
varTuple = ("satu", "dua", "tiga")
varDict = {'nama_produk': "Mouse Logitech B175", 'harga_produk': 100000}

# menampilkan data variabelnya
print(varString)
print(varInt)
print(varFloat)
print(varList)
print(varTuple)
print(varDict)
````
